### (a)

Hessian $ H $ of $ J(\theta) $:

\begin{align*}
\frac {\partial} {\partial \theta_{j}} h_{\theta}(x^{(i)}) & = \frac {\partial} {\partial \theta_{j}} g(\theta^{T} x^{(i)}) \\
& = \frac {\partial} {\partial \theta_{j}} \frac {1} {1 + e^{-\theta^{T} x^{(i)}}} \\
& = - (1+e^{-\theta^{T} x^{(i)}})^{-2} e^{-\theta^{T} x^{(i)}} (- x^{(i)}_{j}) \\
& = (\frac {1} {1 + e^{-\theta^{T} x^{(i)}}} - \frac {1} {(1 + e^{-\theta^{T} x^{(i)}})^{2}}) x^{(i)}_{j} \\
& = g(\theta^{T} x^{(i)})(1 - g(\theta^{T} x^{(i)})) x^{(i)}_{j} \\
& = h_{\theta}(x^{(i)}) (1 - h_{\theta}(x^{(i)})) x^{(i)}_{j}
\end{align*}

\begin{align*}
\frac {\partial} {\partial \theta_{j}} J(\theta) = & \frac {\partial} {\partial \theta_{j}} \big( - \frac {1} {m} \sum_{i=1}^{m} y^{(i)} \log(h_{\theta}(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_{\theta}(x^{(i)})) \big) \\
= & - \frac {1} {m} \sum_{i=1}^{m} y^{(i)} \frac {1} {h_{\theta} (x^{(i)})} h_{\theta} (x^{(i)}) (1 - h_{\theta} (x^{(i)})) x^{(i)}_j \\
& + (1 - y^{(i)}) \frac {1} {1 - h_{\theta} (x^{(i)})} (-1) h_{\theta} (x^{(i)}) (1 - h_{\theta} (x^{(i)})) x^{(i)}_j \\
= & - \frac {1} {m} \sum_{i=1}^{m} (y^{(i)} - h_\theta(x^{(i)})) x^{(i)}_j
\end{align*}

\begin{align*}
\frac {\partial} {\partial \theta_{k}} \frac {\partial} {\partial \theta_{j}} J(\theta) = & \frac {\partial} {\partial \theta_{k}} \big(- \frac {1} {m} \sum_{i=1}^{m} (y^{(i)} - h_\theta(x^{(i)})) x^{(i)}_j \big) \\
= & \frac {1} {m} \sum_{i=1}^{m} h_\theta(x^{(i)}) (1 - h_\theta(x^{(i)})) x_j^{(i)} x_k^{(i)}
\end{align*}

\begin{align*}
H = \frac {1} {m} \sum_{i=1}^{m} h_\theta(x^{(i)}) (1 - h_\theta(x^{(i)})) x^{(i)} (x^{(i)})^T
\end{align*}

show $ z^THz \geq 0 $:

\begin{align*}
z^T H z = & \ z^T \frac {1} {m} \sum_{i=1}^{m} h_{\theta}(x^{(i)}) (1 - h_\theta(x^{(i)})) x^{(i)} (x^{(i)})^T z \\
= & \ \frac {1} {m} \sum_{i=1}^{m} h_\theta(x^{(i)}) (1 - h_\theta(x^{(i)})) z^T x^{(i)} (x^{(i)})^T z \\
= & \ \frac {1} {m} \sum_{i=1}^{m} h_\theta(x^{(i)}) (1 - h_\theta(x^{(i)})) (z^T x^{(i)})^2 \\
\geq & \ 0
\end{align*}

so, $ H \succeq 0 $

### (c)

\begin{align*}
p(y = 1 \vert x; \phi, \mu_0, \mu_1, \Sigma) = & \ \frac {p(x \vert y = 1) p(y = 1)} {p(x)} \\
= & \ \frac {p(x \vert y = 1) p(y = 1)} {p(x \vert y = 1) p(y = 1) + p(x \vert y = 0) p(y = 0)} \\
= & \ \frac {1} {1 + \frac {p(x \vert y = 0) p(y = 0)} {p(x \vert y = 1) p(y = 1)}} \\
= & \ 1 / (1 + \frac {\exp \big(- \frac {1} {2} (x - \mu_0)^T \Sigma^{-1} (x - \mu_0) \big) (1 - \phi)} {\exp \big( - \frac {1} {2} (x - \mu_1)^T \Sigma^{-1} (x - \mu_1) \big) \phi}) \\
= & \ 1 / (1 + \exp \big( - \frac {1} {2} (x - \mu_0)^T \Sigma^{-1} (x - \mu_0) \\
& + \frac {1} {2} (x - \mu_1)^T \Sigma^{-1} (x - \mu_1) \big) \frac {1 - \phi} {\phi}) \\
= & \ 1 / (1 + \exp \Big( - \frac {1} {2} \big( (x - \mu_0)^T \Sigma^{-1} (x - \mu_0) \\
& - (x - \mu_1)^T \Sigma^{-1} (x - \mu_1) \big) \Big) \frac {1 - \phi} {\phi}) \\
= & \ 1 / (1 + \exp \big( - \frac {1} {2} ( x^T\Sigma^{-1}x - x^T\Sigma^{-1}\mu_0 - \mu_0^T\Sigma^{-1}x \\
& + \mu_0^T\Sigma^{-1}\mu_0 -  x^T\Sigma^{-1}x + x^T\Sigma^{-1}\mu_1 + \mu_1^T\Sigma^{-1}x \\
& - \mu_1^T\Sigma^{-1}\mu_1) \big) \frac {1 - \phi} {\phi}) \\
= & \ 1 / (1 + \exp \big( - \frac {1} {2} ( x^T\Sigma^{-1}(\mu_1 - \mu_0) + (\mu_1^T - \mu_0^T)\Sigma^{-1}x \\
& + \mu_0^T\Sigma^{-1}\mu_0 - \mu_1^T\Sigma^{-1}\mu_1 ) \big) \frac {1 - \phi} {\phi}) \\
= & \ 1 / (1 + \exp \big( - \frac {1} {2} ( 2(\mu_1^T - \mu_0^T)\Sigma^{-1}x \\
& + \mu_0^T\Sigma^{-1}\mu_0 - \mu_1^T\Sigma^{-1}\mu_1 ) \big) \exp(\log \frac {1 - \phi} {\phi})) \\
= & \ \frac {1} {1 + \exp \Big( - \big( (\mu_1^T - \mu_0^T)\Sigma^{-1}x + (\frac {1} {2} \mu_0^T\Sigma^{-1}\mu_0 - \frac {1} {2} \mu_1^T\Sigma^{-1} \mu_1 - \log \frac {1 - \phi} {\phi}) \big) \Big)} \\
= & \ \frac {1} {1 + \exp \Big( - \big( (\Sigma^{-1}(\mu_1 - \mu_0))^Tx + (\frac {1} {2} \mu_0^T\Sigma^{-1}\mu_0 - \frac {1} {2} \mu_1^T\Sigma^{-1} \mu_1 - \log \frac {1 - \phi} {\phi}) \big) \Big)}
\end{align*}

so,
\begin{align*}
p(y = 1 \vert x; \phi, \mu_0, \mu_1, \Sigma) = & \ \frac {1} {1 + \exp ( - ( \theta^T x + \theta_0 ) )} \\
\theta = & \ \Sigma^{-1}(\mu_1 - \mu_0) \\
\theta_0 = & \ \frac {1} {2} \mu_0^T\Sigma^{-1}\mu_0 - \frac {1} {2} \mu_1^T\Sigma^{-1} \mu_1 - \log \frac {1 - \phi} {\phi}
\end{align*}

### (d)

\begin{align*}
\ell(\phi, \mu_0, \mu_1, \Sigma) = & \ \log \prod_{i=1}^m p(x^{(i)} \vert y^{(i)}; \mu_0, \mu_1, \Sigma) p(y^{(i)}; \phi) \\
= & \ \log \prod_{i=1}^m \Big( \frac {1} {(2\pi)^{\frac {n} {2}} \vert \Sigma \vert ^{\frac {1} {2}}} \exp \big( - \frac {1} {2} (x^{(i)} - \mu_0)^T \Sigma^{-1} (x^{(i)} - \mu_0) \big) (1 - \phi) \Big)^{1\{y^{(i)} = 0\}} \\
& \ \Big( \frac {1} {(2\pi)^{\frac {n} {2}} \vert \Sigma \vert ^{\frac {1} {2}}} \exp \big( - \frac {1} {2} (x^{(i)} - \mu_1)^T \Sigma^{-1} (x^{(i)} - \mu_1) \big) \phi \Big)^{1\{y^{(i)} = 1\}} \\
= & \ \Sigma_{i = 1}^m 1\{y^{(i)}=0\} \big( \log(2\pi)^{- \frac {n} {2}} + \log \vert \Sigma \vert ^{- \frac {1} {2}} - \frac {1} {2} (x^{(i)} - \mu_0)^T \Sigma^{-1} (x^{(i)} - \mu_0) + \log (1 - \phi) \big) \\
& + \Sigma_{i = 1}^m 1\{y^{(i)}=1\} \big( \log(2\pi)^{- \frac {n} {2}} + \log \vert \Sigma \vert ^{- \frac {1} {2}} - \frac {1} {2} (x^{(i)} - \mu_1)^T \Sigma^{-1} (x^{(i)} - \mu_1) + \log \phi \big) \\
\end{align*}

for $ \phi $:

\begin{align*}
\frac {\partial} {\partial \phi} \ell = & \ \frac {\partial} {\partial \phi} \big( \Sigma^m_{i=1} 1\{y^{(i)}=0\} \log(1 - \phi) + \Sigma^m_{i=1} 1\{y^{(i)}=1\} \log \phi \big) \\
= & \ \Sigma^m_{i=1} 1\{y^{(i)}=0\} \frac {1} {1 - \phi} (-1) + \Sigma^m_{i=1} 1\{y^{(i)}=1\} \frac {1} {\phi} \\
= & \ 0 \\
\Sigma^m_{i=1} 1\{y^{(i)}=0\} \frac {1} {1 - \phi} = & \ \Sigma^m_{i=1} 1\{y^{(i)}=1\} \frac {1} {\phi} \\
\Sigma^m_{i=1} 1\{y^{(i)}=0\} \phi = & \ \Sigma^m_{i=1} 1\{y^{(i)}=1\} (1-\phi) \\
m \phi = & \ \Sigma^m_{i=1} 1 \{ y^{(i)} = 1 \} \\
\phi = & \ \frac {1} {m} \Sigma^m_{i=1} 1 \{ y^{(i)} = 1 \}
\end{align*}

for $ \mu_0 $:

\begin{align*}
\frac {\partial} {\partial \mu_0} \ell = & \ \frac {\partial} {\partial \mu_0} \Sigma^m_{i=1} 1\{ y^{(i)} = 0 \}(- \frac {1} {2} (x^{(i)} - \mu_0)^T \Sigma^{-1} (x^{(i)} - \mu_0)) \\
= & \ \frac {\partial} {\partial \mu_0} \Sigma^m_{i=1} 1\{ y^{(i)} = 0 \} (- \frac {1} {2}) \big( (x^{(i)})^T \Sigma^{-1} x^{(i)} - (x^{(i)})^T \Sigma^{-1} \mu_0 - \mu_0^T \Sigma^{-1} x^{(i)} + \mu_0^T \Sigma^{-1} \mu_0 \big) \\
= & \ \Sigma^m_{i=1} 1\{ y^{(i)}=0 \} \frac {1} {2} \big( (x^{(i)})^T \Sigma^{-1} - \mu_0^T \Sigma^{-1} \big) \\
= & \ \Sigma^m_{i=1} 1\{ y^{(i)}=0 \} \frac {1} {2} \Sigma^{-1} \big( (x^{(i)})^T - \mu_0^T \big) \\
= & \ 0 \\
0 = & \ \Sigma^m_{i=1} 1\{ y^{(i)}=0 \} ( x^{(i)} - \mu_0 ) \\
\Sigma^m_{i=1} 1\{ y^{(i)}=0 \} \mu_0 = & \ \Sigma^m_{i=1} 1\{ y^{(i)}=0 \} x^{(i)} \\
\mu_0 = & \ \frac {\Sigma^m_{i=1} 1\{ y^{(i)}=0 \} x^{(i)}} {\Sigma^m_{i=1} 1\{ y^{(i)}=0 \}}
\end{align*}

for $ \mu_1 $, similar to $ \mu_0 $:

\begin{align*}
\mu_1 = \ \frac { \Sigma^m_{i=1} 1\{ y^{(i)}=1 \} x^{(i)} } { \Sigma^m_{i=1} 1\{ y^{(i)}=1 \} }
\end{align*}

for $ \Sigma $:

\begin{align*}
\frac {\partial} {\partial \Sigma} \ell = & \ \frac {\partial} {\partial \Sigma} \Big( \Sigma_{i = 1}^m 1\{y^{(i)}=0\} \big( \log \vert \Sigma \vert ^{- \frac {1} {2}} - \frac {1} {2} (x^{(i)} - \mu_0)^T \Sigma^{-1} (x^{(i)} - \mu_0) \big) \\
& \ + \Sigma_{i = 1}^m 1\{y^{(i)}=1\} \big( \log \vert \Sigma \vert ^{- \frac {1} {2}} - \frac {1} {2} (x^{(i)} - \mu_1)^T \Sigma^{-1} (x^{(i)} - \mu_1) \big) \Big) \\
= & \ \frac {\partial} {\partial \Sigma} \Sigma^m_{i=1} \big( \log \vert \Sigma \vert ^{- \frac {1} {2}} - \frac {1} {2} (x^{(i)} - \mu_{y^{(i)}})^T \Sigma^{-1} (x^{(i)} - \mu_{y^{(i)}}) \big) \\
= & \ \frac {\partial} {\partial \Sigma} \Sigma^m_{i=1} \big( - \frac {1} {2} \log \vert \Sigma \vert - \frac {1} {2} \Sigma^{-1} (x^{(i)} - \mu_{y^{(i)}}) (x^{(i)} - \mu_{y^{(i)}})^T \big) \\
= & \ \Sigma^m_{i=1} \big( - \frac {1} {2} \frac {1} {\vert \Sigma \vert} + \frac {1} {2} \Sigma^{-2} (x^{(i)} - \mu_{y^{(i)}}) (x^{(i)} - \mu_{y^{(i)}})^T \big) \\
= & \ 0 \\
\Sigma^m_{i=1} \frac {1} {\vert \Sigma \vert} = & \ \Sigma^m_{i=1} \Sigma^{-2} (x^{(i)} - \mu_{y^{(i)}}) (x^{(i)} - \mu_{y^{(i)}})^T \\
m \Sigma = & \ \Sigma^m_{i=1} (x^{(i)} - \mu_{y^{(i)}}) (x^{(i)} - \mu_{y^{(i)}})^T \\
\Sigma = & \ \frac {1} {m} \Sigma^m_{i=1} (x^{(i)} - \mu_{y^{(i)}}) (x^{(i)} - \mu_{y^{(i)}})^T
\end{align*}

```python
import matplotlib.pyplot as plt
```

```python
# modify plot in util
def myplot(x, y, theta_a, theta_a_name, theta_b, theta_b_name,
           title=None, correction=1.0):
    # Plot dataset
    plt.figure()
    plt.plot(x[y == 1, -2], x[y == 1, -1], 'bx', linewidth=2)
    plt.plot(x[y == 0, -2], x[y == 0, -1], 'go', linewidth=2)

    # Plot decision boundary (found by solving for theta^T x = 0)
    margin1 = (max(x[:, -2]) - min(x[:, -2]))*0.2
    margin2 = (max(x[:, -1]) - min(x[:, -1]))*0.2
    
    x1 = np.arange(min(x[:, -2])-margin1, max(x[:, -2])+margin1, 0.01)
    
    x2_a = -(theta_a[0] / theta_a[2] * correction \
             + theta_a[1] / theta_a[2] * x1)
    plt.plot(x1, x2_a, c='red', label=theta_a_name, linewidth=2)
    
    x2_b = -(theta_b[0] / theta_b[2] * correction \
             + theta_b[1] / theta_b[2] * x1)
    plt.plot(x1, x2_b, c='yellow', label=theta_b_name, linewidth=2)
    
    plt.xlim(x[:, -2].min()-margin1, x[:, -2].max()+margin1)
    plt.ylim(x[:, -1].min()-margin2, x[:, -1].max()+margin2)
    
    # show legend
    plt.legend(loc='best')
    
    # show title
    if title: plt.title(title)

    # Add labels
    plt.xlabel('x1')
    plt.ylabel('x2')
```

```python
myplot(ds1_x_train, ds1_y_train, clf_ds1.theta, 'Logistic Regression',
       gda_ds1.theta, 'GDA', 'Dataset1 train')
```

```python
myplot(ds1_x_valid, ds1_y_valid, clf_ds1.theta, 'Logistic Regression',
       gda_ds1.theta, 'GDA', 'Dataset1 valid')
```
